```python
def check_and_label(df, column_name, predefined_list):
    # Function to check if any word from the predefined list exists in a string
    predefined_list=set(predefined_list)
#     df['message']=df['message'].astype(str)
    def check_words_exist(text):
        for word in predefined_list:
            if word.lower() in text.lower().split():
                return 0
        return 1

    # Apply the check_words_exist function to the specified column
    df['offensive'] = df[column_name].apply(check_words_exist)

    return df
```

```python
def remove_double_quotes(input_string):
    # Define the regex patterns to match double quotes with or without spaces around them
    pattern1 = r'\s*"'
    pattern2 = r'"\s*'
    
    # Use re.sub() to replace occurrences of the patterns with an empty string
    result_string = re.sub(pattern1, '', input_string)
    result_string = re.sub(pattern2, '', result_string)
    
    return result_string
```

```python
def remove_question_marks(text):
    # Define the regular expression pattern
    pattern = r'\?'

    # Use the re.sub() function to remove question marks
    cleaned_text = re.sub(pattern, '', text)

    return cleaned_text
```

```python
def remove_links(text):
    # Define the regular expression pattern for URLs
    pattern = r'https?://\S+|www\.\S+'

    # Use the re.sub() function to remove links
    cleaned_text = re.sub(pattern, '', text)

    return cleaned_text
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
import string
stop_words = nltk.corpus.stopwords.words('english')
ps = PorterStemmer()
```

```python
def remove_stop_words(sentence):
    words = sentence.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
def process_alphanumeric(input_string):
    words = input_string.split()  # Split the string into words using spaces
    alphanumeric_words = [word for word in words if word.isalnum()]  # Filter alphanumeric words
    
    result = ' '.join(alphanumeric_words)  # Join the alphanumeric words back into a string
    return result
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    
    y = []
    
    for i in text:
        if i not in stop_words and i not in string.punctuation:
            y.append(i)
            
    text = y[:]
    y.clear()
    
    for i in text:
        y.append(ps.stem(i))
    
            
    return ' '.join(y)
```

```python
data['message']=data['message'].apply(transform_text)
```
